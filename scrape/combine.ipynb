```python
df2=pd.read_csv('houses_data.csv')
```

```python
print(sum(df2.duplicated()))
df2.drop_duplicates(inplace=True)
```

```python
sum(combined_df.duplicated())
```

```python
combined_df.drop_duplicates(inplace=True)
print(combined_df.shape)
```
